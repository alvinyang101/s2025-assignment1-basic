```python
%load_ext autoreload
%autoreload 2
from pathlib import Path
import cProfile, pstats
import pickle

from ece496b_basics.bpe import train_bpe

DATA_PATH = Path("../data").resolve()
OUTPUT_PATH = Path("outputs").resolve()
OUTPUT_PATH.mkdir(parents=True, exist_ok=True)
input_path = DATA_PATH / "TinyStoriesV2-GPT4-train.txt"

def train_bpe_tinystories():
    vocab, merged = train_bpe(
    input_path=input_path,
    vocab_size=10000,
    special_tokens=["<|endoftext|>"],
    )
    return (vocab, merged)


with cProfile.Profile() as pr:
    data = train_bpe_tinystories()

with open(OUTPUT_PATH / "tinystories_vocab.pkl", "wb") as file:
    pickle.dump(data[0], file)
with open(OUTPUT_PATH / "tinystories_merges.pkl", "wb") as file:
    pickle.dump(data[1], file)
```

```python
stats = pstats.Stats(pr)
stats.sort_stats(pstats.SortKey.TIME)
stats.print_stats()
```
