```python
%load_ext autoreload
%autoreload 2
from pathlib import Path
import pickle

from ece496b_basics.bpe import train_bpe

DATA_PATH = Path("../data").resolve()
OUTPUT_PATH = Path("outputs").resolve()
OUTPUT_PATH.mkdir(parents=True, exist_ok=True)
input_path = DATA_PATH / "owt_train.txt"

def train_bpe_openweb():
    vocab, merged = train_bpe(
    input_path=input_path,
    vocab_size=32000,
    special_tokens=["<|endoftext|>"],
    )
    return (vocab, merged)


data = train_bpe_openweb()

with open(OUTPUT_PATH / "owt_vocab.pkl", "wb") as file:
    pickle.dump(data[0], file)
with open(OUTPUT_PATH / "owt_merges.pkl", "wb") as file:
    pickle.dump(data[1], file)
```

```python
max_val = ""
for key, val in data[0].items():
    print(val)
    if len(val) > len(max_val):
        max_val = val

print(max_val)
```
